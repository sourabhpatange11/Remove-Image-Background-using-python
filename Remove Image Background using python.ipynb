```python
import rembg
from PIL import Image
from matplotlib import pyplot as plt

def remove_background(input_image_path, output_image_path):
    with open(input_image_path, "rb") as input_file, open(output_image_path, "wb") as output_file:
        input_image = rembg.remove(input_file.read())
        output_file.write(input_image)

input_image_path = "C:\\Users\\soura\\Downloads\\sample.jpg"  # Replace this with your input image path
output_image_path = "C:\\Users\\soura\\Downloads\\background_remove.png"  # Replace this with the desired output image path

remove_background(input_image_path, output_image_path)

# Load original and background removed images using PIL
original_image = Image.open(input_image_path)
background_removed_image = Image.open(output_image_path)

# Display images using Matplotlib
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Original Image
axes[0].imshow(original_image)
axes[0].set_title('Original Image')
axes[0].axis('off')

# Background Removed Image
axes[1].imshow(background_removed_image)
axes[1].set_title('Background Removed Image')
axes[1].axis('off')

plt.tight_layout()
plt.show()
```
